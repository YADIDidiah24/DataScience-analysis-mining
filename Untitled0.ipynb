```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
data = {
    'Characteristic': ['GET Score', 'Need For Achievement', 'Need For Autonomy', 'Creative Tendency', 'Calculated Risk Taking', 'Locus Of Control'],
    'Test 1': [85, 92, 83, 100, 75, 75],
    'Test 2': [93, 92, 100, 100, 83, 92]
}

# Create DataFrame
df = pd.DataFrame(data)

# Melt DataFrame
df_melted = df.melt(id_vars='Characteristic', var_name='Test', value_name='Score')

# Plot

custom_palette = sns.color_palette("hsv", 2)  # hsv palette for bright colors
sns.barplot(data=df_melted, x='Characteristic', y='Score', hue='Test', palette=custom_palette)

# Add data labels
for index, row in df_melted.iterrows():
    plt.text(row.name, row.Score, str(row.Score), color='black', ha="center")

plt.title('Comparison of Test Scores')
plt.xlabel('Characteristic')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Test', loc='upper right')
plt.show()
```
